```python
#Thymine - for
list_TF = []
for k in range(151):
    thy = 0
    for sample in letterlist_for:
        if sample[k:k+1]=="T":
            thy += 1
    per = thy/n
    list_TF.append(per)
#Adenine - for
list_AF = []
for k in range(151):
    ade = 0
    for sample in letterlist_for:
        if sample[k:k+1]=="A":
            ade += 1
    per = ade/n
    list_AF.append(per)
#Cytosine - for
list_CF = []
for k in range(151):
    cyt = 0
    for sample in letterlist_for:
        if sample[k:k+1]=="C":
            cyt += 1
    per = cyt/n
    list_CF.append(per)
#Guanine - for
list_GF = []
for k in range(151):
    gua = 0
    for sample in letterlist_for:
        if sample[k:k+1]=="G":
            gua += 1
    per = gua/n
    list_GF.append(per)
list_allF = list(map(sum, zip(list_TF,list_AF,list_CF,list_GF))) #for control
```
